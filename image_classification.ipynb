```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
img = cv2.imread('0_bear.JPEG')

# 256 * 3 = 768
x_train = np.empty((3, 768))
rgb_pixel = []

for input_num in range(1):
  # 畫出 RGB 三種顏色的分佈圖
  color = ('b','g','r')
  for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0, 256])
    #plt.plot(histr, color = col)
    plt.hist(histr.ravel(), 256, [0, 256])
    plt.xlim([0, 256])
    
    # 顯示b, r, g的1維數值
    print("-- ", col, " --")
    print(histr.ravel())
    rgb_pixel.extend(histr.ravel())
plt.show()
x_train[0:1] = rgb_pixel[:]
```
