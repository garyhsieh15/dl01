setp1. 讀取圖片, 擷取圖片特徵, 然後將特徵像素數值, 儲存成0 ~ 255個種類

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```
#img = cv2.imread('0_bear.JPEG')

xfiles = os.listdir("./")
xfiles_len = len(xfiles)

train_sample = xfiles_len
train_label  = xfiles_len
# 256 * 3 = 768
pixel_num    = 768

x_train = np.empty((xfiles_len, pixel_num))
y_train = np.empty(xfiles_len)

print("show len xfiles: ", xfiles_len)

xfile_idx = 0
for xfile_name in xfiles:
    ext = xfile_name.split('_')
    #print("file_name and ext info: ", file_name, ext)
    if xfile_idx >= xfiles_len:
      break
    if ext[-1] == "bear.JPEG":
        y_train[xfile_idx] = 0
    else:
        y_train[xfile_idx] = 1
    xfile_idx += 1


# for loop via number of pic
for dataset_num in range(xfiles_len):
#for dataset_num in range(1):
  print("show pic name: ", xfiles[dataset_num])
  img = cv2.imread(xfiles[dataset_num])
  #img = cv2.imread("111_leopard.JPEG")
  
  rgb_pixel = []
  # 畫出 RGB 三種顏色的分佈圖
  color = ('b','g','r')
  for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0, 256])
    #plt.plot(histr, color = col)
    #plt.hist(histr.ravel(), 256, [0, 256])
    #plt.xlim([0, 256])
    
    # 顯示b, r, g的1維數值
    #print("-- ", col, " --")
    #print(histr.ravel())
    rgb_pixel.extend(histr.ravel())
  x_train[dataset_num:dataset_num + 1] = rgb_pixel[:] / max(rgb_pixel[:])
#plt.show()
```

step2. 訓練資料

```
# sigmoid(x) function
def sigmoid(x):
  #print("show sigmoid x value: ", x)
  #print("show sigmoid value: ", 1/(1 + np.exp(-x)))
  return 1/(1 + np.exp(-x))

# activation function
def activation(x, w, b):
  #print("show np.dot: ", np.dot(x, w))
  return sigmoid(np.dot(x, w) + b)

# updata weight and bias value
def update(x, y, w, b, eta):
  y_pred = activation(x, w, b)
  #print("show y_pred: ", y_pred)
  a = (y_pred - y_train) * y_pred * (1 - y_pred)
  for i in range(pixel_num):
  #for i in range(4):
    w[i] -= eta * 1/float(len(y)) * np.sum(a * x[:, i])
  b -= eta * 1/float(len(y)) * np.sum(a)
    #print("show a*x[0:1, i]: ", a*x[0:1, i])
  return w, b
```

```
# initial weight value
weights = np.ones(pixel_num) / 10
bias = np.ones(1) / 10
for _ in range(1000):
  weights, bias = update(x_train, y_train, weights, bias, eta = 0.1)
  #print("weight = %f, bias = %f" % (weights, bias))
print("weights = ", weights, "bias = ", bias)
```
